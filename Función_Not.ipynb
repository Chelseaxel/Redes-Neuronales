```python
# Operación lógica NOT con una neurona
import numpy as np

# Entradas (solo un valor por vez)
X = np.array([[0], [1]])
Y = np.array([1, 0])

# Pesos y sesgo iniciales
w = np.random.rand(1)
b = np.random.rand(1)
alpha = 0.1

def step(x):
    return 1 if x >= 0 else 0

# Entrenamiento
for epoch in range(10):
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = step(z)
        error = Y[i] - y_pred
        w += alpha * error * X[i]
        b += alpha * error

# Pruebas del perceptrón NOT
print("Peso final:", w)
print("Sesgo final:", b)

print("\nResultados del perceptrón NOT:")
for i in range(len(X)):
    z = np.dot(X[i], w) + b
    print(f"{X[i]} → {step(z)}")
```

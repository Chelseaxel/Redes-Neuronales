```python
# Operación lógica OR con una neurona (Perceptrón)
import numpy as np

# Datos de entrada (X) y salidas esperadas (Y)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
Y = np.array([0, 1, 1, 1])

# Inicialización de pesos y sesgo
w = np.random.rand(2)
b = np.random.rand(1)
alpha = 0.1  # tasa de aprendizaje

# Función de activación
def step(x):
    return 1 if x >= 0 else 0

# Entrenamiento
for epoch in range(10):
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = step(z)
        error = Y[i] - y_pred
        w += alpha * error * X[i]
        b += alpha * error

# Pruebas del perceptrón OR
print("Pesos finales:", w)
print("Sesgo final:", b)

print("\nResultados del perceptrón OR:")
for i in range(len(X)):
    z = np.dot(X[i], w) + b
    print(f"{X[i]} → {step(z)}")
```
